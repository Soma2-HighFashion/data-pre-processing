```python
import os
from SimpleCV import *
init_options_handler.enable_notebook()
```

```python
def preprocessing(image, dest):
    crop_image = image.crop(image.width*0.25, 0, image.width*0.5, image.height)
    
#     diff = int((crop_image.height - crop_image.width)/2)
#     right_white_image = crop_image.embiggen((crop_image.width+diff, crop_image.height), Color.WHEAT, (0, 0))
    
#     square_image = right_white_image.embiggen((right_white_image.width+diff, right_white_image.height), Color.WHEAT, (diff, 0))
#     left10 = right_white_image.embiggen((right_white_image.width+diff, right_white_image.height), Color.WHEAT, (diff+10, 0))
#     right10 = right_white_image.embiggen((right_white_image.width+diff, right_white_image.height), Color.WHEAT, (diff-10, 0))
    
    save84x256(crop_image, dest + "_01.jpg")
    save84x256(crop_image.flipHorizontal(), dest + "_02.jpg")
#     save_128x128(left10, dest + "_03.jpg")
#     save_128x128(left10.flipHorizontal(), dest + "_04.jpg")
#     save_128x128(right10, dest + "_05.jpg")
#     save_128x128(right10.flipHorizontal(), dest + "_06.jpg")

def save84x256(image, dest):
    image.adaptiveScale((84, 256)).save(dest)
```

```python
src_dir = "./../category_data/img_street/"
img_data = os.listdir(src_dir)
dest_dir = "./../output/img_street/"

for i in xrange(len(img_data)): 
    try:
        image_path = src_dir + img_data[i]
        image = Image(image_path)
        preprocessing(image, dest_dir+"{:05d}".format((i+1)))
    except Exception as e:
        print e
```
