```python
m_count = 5
b_count = 3

def preprocessing_1(image2, dest):
    image2 = image2.erode()
    image2 = image2.crop(image2.width*0.25, 0, image2.width*0.5, image2.height).adaptiveScale((42, 128))
    image2 = image2.dilate()
    save_img(image2, dest)
    return

def preprocessing_2(image2, dest):
    image2 = image2.dilate()
    image2 = image2.erode()
    image2 = image2.crop(image2.width*0.25, 0, image2.width*0.5, image2.height).adaptiveScale((42, 128))
    save_img(image2, dest)
    return

def preprocessing_3(image2, dest):
    image2 = smooth_median(image2, m_count)
    image2 = image2.crop(image2.width*0.25, 0, image2.width*0.5, image2.height).adaptiveScale((42, 128))
    image2 = smooth_bilateral(image2, b_count)
    save_img(image2, dest)
    return

def preprocessing_4(image2, dest):
    image2 = smooth_bilateral(image2, b_count)
    image2 = image2.crop(image2.width*0.25, 0, image2.width*0.5, image2.height).adaptiveScale((42, 128))
    save_img(image2, dest)
    return

def preprocessing_5(image2, dest):
    image2 = smooth_median(image2, m_count)
    image2 = smooth_bilateral(image2, b_count)
    image2 = image2.crop(image2.width*0.25, 0, image2.width*0.5, image2.height).adaptiveScale((42, 128))
    save_img(image2, dest)
    return
    
def save_img(image, dest):
    image.save(dest + "_01.jpg")
    image.flipHorizontal().save(dest + "_02.jpg")
    return 

def smooth_gausian(image, n):
    for i in range(n):
        image = image.smooth(algorithm_name='gausian')
    return image

def smooth_median(image, n):
    for i in range(n):
        image = image.smooth(algorithm_name='median')
    return image

def smooth_bilateral(image, n):
    for i in range(n):
        image = image.smooth(algorithm_name='bilateral')
    return image

def dilate_then_erode(image):
    image = image.dilate()
    image = image.erode()
    return image
```

```python
child_dir = "male/img_classic/"
src_dir = "./../img_analysis_data2/" + child_dir
img_data = os.listdir(src_dir)
dest_dir = "./../output/" + child_dir

for i in xrange(len(img_data)): 
    try:
        image_path = src_dir + img_data[i]
        image = Image(image_path)
        
        preprocessing_1(image, dest_dir+"{:05d}_1".format((i+1)))
        preprocessing_2(image, dest_dir+"{:05d}_2".format((i+1)))
        preprocessing_3(image, dest_dir+"{:05d}_3".format((i+1)))
        preprocessing_4(image, dest_dir+"{:05d}_4".format((i+1)))
        preprocessing_5(image, dest_dir+"{:05d}_5".format((i+1)))
        
    except Exception as e:
        print e
```
