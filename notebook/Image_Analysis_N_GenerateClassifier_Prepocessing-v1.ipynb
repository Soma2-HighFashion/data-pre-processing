```python
import os
from SimpleCV import *
init_options_handler.enable_notebook()
```

```python
def preprocessing(image, dest):
    image = image.crop(image.width*0.25, 0, image.width*0.5, image.height)
    image = image.adaptiveScale((42, 128))
    image = image.dilate()
    image = image.erode()
    image.save(dest + "_01.jpg")
    image.flipHorizontal().save(dest + "_02.jpg")
    return 

def smooth_gausian(image, n):
    for i in range(n):
        image = image.smooth(algorithm_name='gausian')
    return image

def smooth_median(image, n):
    for i in range(n):
        image = image.smooth(algorithm_name='median')
    return image

def smooth_bilateral(image, n):
    for i in range(n):
        image = image.smooth(algorithm_name='bilateral')
    return image

def dilate_then_erode(image):
    image = image.dilate()
    image = image.erode()
    return image
```

```python
child_dir = "female/img_unique/"
src_dir = "./../img_analysis_data/" + child_dir
img_data = os.listdir(src_dir)
dest_dir = "./../output/" + child_dir

for i in xrange(len(img_data)): 
    try:
        image_path = src_dir + img_data[i]
        image = Image(image_path)
        preprocessing(image, dest_dir+"{:05d}".format((i+1)))
    except Exception as e:
        print e
```
